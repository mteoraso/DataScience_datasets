# **Modelo no lineal para prediccion de solubilidad de moleculas**

La solubilidad es la capacidad de una sustancia, denominada soluto, de disolverse en otra llamada disolvente; es decir, es la cantidad máxima de un soluto que un solvente puede recibir en determinadas condiciones ambientales.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.feature_selection import mutual_info_regression

df=pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

```python
df.columns
```

# **Analisis Exploratorio de Datos**

```python
df.describe()
```

# **Ingenieria de caracteristicas**

```python
# Se definen todas las caracteristicas
X=df.drop(['logS'], axis=1)
X
```

```python
# Se establece la variable objetivo
Y=df.iloc[:,-1]
Y
```

```python
# Histogramas de caracteristicas

fig = plt.figure()
fig = plt.figure(figsize=(8, 10))
ax = fig.subplots(5)

ax[0].hist(df["MolLogP"])
ax[0].set_xlabel('MolLogP')
ax[0].set_ylabel('Frecuencia')

ax[1].hist(df["MolWt"])
ax[1].set_xlabel('MolWt')
ax[1].set_ylabel('Frecuencia')

ax[2].hist(df["NumRotatableBonds"])
ax[2].set_xlabel('NumRotatableBonds')
ax[2].set_ylabel('Frecuencia')

ax[3].hist(df["AromaticProportion"])
ax[3].set_xlabel('AromaticProportion')
ax[3].set_ylabel('Frecuencia')

ax[4].hist(df["logS"])
ax[4].set_xlabel('logS')
ax[4].set_ylabel('Frecuencia')

fig.tight_layout()
plt.show()
```

```python
# Evaluando informacion mutua de caracteristicas con respecto al objetivo
mi_scores = mutual_info_regression(X, Y)
mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
mi_scores = mi_scores.sort_values(ascending=False)
mi_scores
```

```python
scores = mi_scores.sort_values(ascending=True)
width = np.arange(len(scores))
ticks = list(scores.index)
plt.barh(width, scores)
plt.yticks(width, ticks)
plt.title("Mutual Information Scores")
plt.figure(dpi=100, figsize=(8, 5))
```

```python
# Analisis de relaciones entre caracteristicas

fig = plt.figure()
fig = plt.figure(figsize=(8, 10))
ax = fig.subplots(4)

ax[0].scatter(x=df["MolLogP"], y=df["logS"], alpha= 0.8)
ax[0].set_xlabel('MolLogP')
ax[0].set_ylabel('logS')

ax[1].scatter(x=df["MolWt"], y=df["logS"], alpha= 0.8)
ax[1].set_xlabel('MolWt')
ax[1].set_ylabel('logS')

ax[2].scatter(x=df["NumRotatableBonds"], y=df["logS"], alpha= 0.8)
ax[2].set_xlabel('NumRotatableBonds')
ax[2].set_ylabel('logS')

ax[3].scatter(x=df["AromaticProportion"], y=df["logS"], alpha= 0.8)
ax[3].set_xlabel('AromaticProportion')
ax[3].set_ylabel('logS')

fig.tight_layout()
plt.show()
```

```python
# Crear una matriz de correlación para el conjunto de datos "tips"
correlation_matrix = df.corr(numeric_only=True)

# Crear un mapa de calor para visualizar la matriz de correlación
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", center=0)
```

# **Entrenamiento y pruebas de modelos**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

Construcción de modelo de regresión lineal

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train, Y_train)
```

```python
#Coeficientes de Y = mx + b
lr.coef_
```

```python
#Este es el valor donde corta el eje Y (en X=0)
lr.intercept_
```

Haciendo predicciones

```python
y_lr_train_pred=lr.predict(X_train)
y_lr_test_pred=lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

```python
#Prediciones, se deben ingresar 4 valores, correspondientes a las otras columnas del DataFrame original

vr_pred_logS = lr.predict([[2.91890, 187.375, 1.0, 0.000000]])
print(int(vr_pred_logS))
```

Evaluando performance del modelo

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse=mean_squared_error(Y_train, y_lr_train_pred)
lr_train_r2=r2_score(Y_train, y_lr_train_pred)

lr_test_mse=mean_squared_error(Y_test, y_lr_test_pred)
lr_test_r2=r2_score(Y_test, y_lr_test_pred)
```

```python
print("El error medio cuadratico de los datos de entrenamiento con respecto a los predecidos es: ", lr_train_mse)
print("El R2 de los datos de entrenamiento con respecto a los predecidos es: ",lr_train_r2)
print("El error medio cuadratico de los datos de prueba con respecto a los predecidos es: ",lr_test_mse)
print("El R2 de los datos de prueba con respecto a los predecidos es: ",lr_test_r2)
```

```python
lr_results=pd.DataFrame(['Linear regression - base model',
                         lr_train_mse,
                         lr_train_r2,
                         lr_test_mse,
                         lr_test_r2]).transpose()

lr_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

Construccion de modelo Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(max_depth=2, random_state=42)
rf.fit(X_train, Y_train)
```

Haciendo predicciones a partir de datos de entrenamiento

```python
y_rf_train_pred=rf.predict(X_train)
y_rf_test_pred=rf.predict(X_test)
```

Evaluando performance del modelo

```python
rf_train_mse=mean_squared_error(Y_train, y_rf_train_pred)
rf_train_r2=r2_score(Y_train, y_rf_train_pred)

rf_test_mse=mean_squared_error(Y_test, y_rf_test_pred)
rf_test_r2=r2_score(Y_test, y_rf_test_pred)
```

```python
rf_results=pd.DataFrame(['Random Forest - base model',
                         rf_train_mse,
                         rf_train_r2,
                         rf_test_mse,
                         rf_test_r2]).transpose()

rf_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

```python
# Evaluacion de modelo de regresion lineal reduciendo caracteristicas

# creacion de dataframe con 2 caracteristicas
X_mol_1=df.drop(['logS','NumRotatableBonds',"AromaticProportion"], axis=1)

# Obteniendo valores de entrenamiento y prueba
X_mol_1_train, X_mol_1_test, Y_mol_1_train, Y_mol_1_test = train_test_split(X_mol_1, Y, test_size=0.2, random_state=42)

# Ajustando a modelo de regresion lineal
lr.fit(X_mol_1_train, Y_mol_1_train)

y_mol_1_lr_train_pred=lr.predict(X_mol_1_train)
y_mol_1_lr_test_pred=lr.predict(X_mol_1_test)

# Calculando el error cuadrado promedio
lr_mol_1_train_mse=mean_squared_error(Y_mol_1_train, y_mol_1_lr_train_pred)
lr_mol_1_train_r2=r2_score(Y_mol_1_train, y_mol_1_lr_train_pred)

lr_mol_1_test_mse=mean_squared_error(Y_mol_1_test, y_mol_1_lr_test_pred)
lr_mol_1_test_r2=r2_score(Y_mol_1_test, y_mol_1_lr_test_pred)

# Dataframe para presentacion de resultados de performance del modelo
lr_mol_1_results=pd.DataFrame(['Linear Regression - 2 features',
                         lr_mol_1_train_mse,
                         lr_mol_1_train_r2,
                         lr_mol_1_test_mse,
                         lr_mol_1_test_r2]).transpose()

lr_mol_1_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_mol_1_results
```

```python
# Evaluacion de modelo de regresion lineal reduciendo caracteristicas

# Creacion de dataframe con 1 caracteristica
X_mol_2=df.drop(['logS','NumRotatableBonds',"AromaticProportion",'MolWt'], axis=1)

# Obteniendo valores de entrenamiento y prueba
X_mol_2_train, X_mol_2_test, Y_mol_2_train, Y_mol_2_test = train_test_split(X_mol_2, Y, test_size=0.2, random_state=42)

# Ajustando a modelo de regresion lineal
lr.fit(X_mol_2_train, Y_mol_2_train)

y_mol_2_lr_train_pred=lr.predict(X_mol_2_train)
y_mol_2_lr_test_pred=lr.predict(X_mol_2_test)

# Calculando el error cuadrado promedio
lr_mol_2_train_mse=mean_squared_error(Y_mol_2_train, y_mol_2_lr_train_pred)
lr_mol_2_train_r2=r2_score(Y_mol_2_train, y_mol_2_lr_train_pred)

lr_mol_2_test_mse=mean_squared_error(Y_mol_2_test, y_mol_2_lr_test_pred)
lr_mol_2_test_r2=r2_score(Y_mol_2_test, y_mol_2_lr_test_pred)

# Dataframe para presentacion de resultados de performance del modelo
lr_mol_2_results=pd.DataFrame(['Linear Regression - 1 feature',
                         lr_mol_2_train_mse,
                         lr_mol_2_train_r2,
                         lr_mol_2_test_mse,
                         lr_mol_2_test_r2]).transpose()

lr_mol_2_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_mol_2_results
```

```python
# Evaluacion de modelo de random forest reduciendo caracteristicas

# Creacion de dataframe con 3 caracteristicas
X_mol_rf_1=df.drop(['logS','NumRotatableBonds'], axis=1)

# Obteniendo valores de entrenamiento y prueba
X_mol_rf_1_train, X_mol_rf_1_test, Y_mol_rf_1_train, Y_mol_rf_1_test = train_test_split(X_mol_rf_1, Y, test_size=0.2, random_state=42)

# Ajustando a Random Forest Regressor
rf_1=RandomForestRegressor(max_depth=4, random_state=42)

# Ajustando a modelo de random forest
rf_1.fit(X_mol_rf_1_train, Y_mol_rf_1_train)

y_mol_rf_1_train_pred=rf_1.predict(X_mol_rf_1_train)
y_mol_rf_1_test_pred=rf_1.predict(X_mol_rf_1_test)

# Calculando el error cuadrado promedio
rf_1_mol_train_mse=mean_squared_error(Y_mol_rf_1_train, y_mol_rf_1_train_pred)
rf_1_mol_train_r2=r2_score(Y_mol_rf_1_train, y_mol_rf_1_train_pred)

rf_1_mol_test_mse=mean_squared_error(Y_mol_rf_1_test, y_mol_rf_1_test_pred)
rf_1_mol_test_r2=r2_score(Y_mol_rf_1_test, y_mol_rf_1_test_pred)

# Dataframe para presentacion de resultados de performance del modelo
rf_1_mol_results=pd.DataFrame(['Random Forest - 3 features',
                         rf_1_mol_train_mse,
                         rf_1_mol_train_r2,
                         rf_1_mol_test_mse,
                         rf_1_mol_test_r2]).transpose()

rf_1_mol_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_1_mol_results
```

```python
# Evaluacion de modelo de random forest reduciendo caracteristicas

# Creacion de Dataframe con 2 caracteristicas
X_mol_rf_2=df.drop(['logS','NumRotatableBonds','AromaticProportion'], axis=1)

# Obteniendo valores de entrenamiento y prueba
X_mol_rf_2_train, X_mol_rf_2_test, Y_mol_rf_2_train, Y_mol_rf_2_test = train_test_split(X_mol_rf_2, Y, test_size=0.2, random_state=42)

# Ajustando a modelo Random Forest Regressor
rf_2=RandomForestRegressor(max_depth=4, random_state=42)

# Ajustando a modelo de random forest
rf_2.fit(X_mol_rf_2_train, Y_mol_rf_2_train)

y_mol_rf_2_train_pred=rf_2.predict(X_mol_rf_2_train)
y_mol_rf_2_test_pred=rf_2.predict(X_mol_rf_2_test)

# Calculando el error cuadrado promedio
rf_2_mol_train_mse=mean_squared_error(Y_mol_rf_2_train, y_mol_rf_2_train_pred)
rf_2_mol_train_r2=r2_score(Y_mol_rf_2_train, y_mol_rf_2_train_pred)

rf_2_mol_test_mse=mean_squared_error(Y_mol_rf_2_test, y_mol_rf_2_test_pred)
rf_2_mol_test_r2=r2_score(Y_mol_rf_2_test, y_mol_rf_2_test_pred)

# Dataframe para presentacion de resultados de performance del modelo
rf_2_mol_results=pd.DataFrame(['Random Forest - 2 features',
                         rf_2_mol_train_mse,
                         rf_2_mol_train_r2,
                         rf_2_mol_test_mse,
                         rf_2_mol_test_r2]).transpose()

rf_2_mol_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_2_mol_results
```

```python
#Comparacion de modelos
indices = [("Linear regression","Base model"), ("Linear regression","2 Features"),("Linear regression","1 Features"),
             ("Random Forest","Base model"), ("Random Forest","3 Features"),("Random Forest","2 Features")]

index = pd.MultiIndex.from_tuples(indices)

# Crear un DataFrame vacío con el multi índice
df_multi_index = pd.DataFrame(index=index, columns=['Training MSE', 'Training R2', 'Test MSE', 'Test R2'])

# Rellenar las celdas con los valores de los otros DataFrames
df_multi_index.loc[('Linear regression', 'Base model')] = lr_results.iloc[0, 1:].values
df_multi_index.loc[('Linear regression', '2 Features')] = lr_mol_1_results.iloc[0, 1:].values
df_multi_index.loc[('Linear regression', '1 Features')] = lr_mol_2_results.iloc[0, 1:].values
df_multi_index.loc[('Random Forest', 'Base model')] = rf_results.iloc[0, 1:].values
df_multi_index.loc[('Random Forest', '3 Features')] = rf_1_mol_results.iloc[0, 1:].values
df_multi_index.loc[('Random Forest', '2 Features')] = rf_2_mol_results.iloc[0, 1:].values
display(df_multi_index)
```
