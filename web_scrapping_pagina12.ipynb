```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(p12.text, 'lxml')
```

```python
type(s)
```

```python
print(s.prettify())
```

```python
s.find('ul')
```

```python
s.find('ul', {'class': 'horizontal-list main-sections hide-on-dropdown'})
```

```python
s.find('ul', {'class': 'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
secciones = s.find('ul', {'class': 'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
secciones
```

```python
seccion = secciones[0]
```

```python
seccion
```

```python
seccion.find('a')
```

```python
seccion.a.get('href')
```

```python
seccion.a.get_text()
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
```

```python
links_secciones
```

```python
sec = requests.get(links_secciones[0])
```

```python
sec
```

```python
sec.status_code
```

```python
s_seccion = BeautifulSoup(sec.text, 'lxml')
```

```python
print(s_seccion.prettify())
```

```python
featured_article = s_seccion.find('div', attrs={'class':'main-content'})
featured_article
```

```python
featured_article.a.get('href')
```

```python
featured_article.find('ul')
```
